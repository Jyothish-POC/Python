### Embedding Techniques

```python
import os
from dotenv import load_dotenv
load_dotenv() #Load enenvironment variable from env file
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "This is a tutorial on OPENAI embeddings."
query_result = embeddings.embed_query(text)
```

```python
query_result
```

```python
len(query_result)  # Check the length of the embedding vector
```

```python
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-large", dimension=1024)
```

```python
embeddings_1024
```

```python
text = "This is a tutorial on OPENAI embeddings."
query_result = embeddings_1024.embed_query(text)
len(query_result)  # Check the length of the embedding vector
```
